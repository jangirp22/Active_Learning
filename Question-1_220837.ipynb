```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

print("The Iris dataset:")
print(iris.head())

iris.hist()
plt.suptitle('Histograms of Iris Dataset Columns')
plt.show()

sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.title('Scatter Plots of Iris Dataset')
plt.show()

numerical_columns = iris.select_dtypes(include='number').columns

mean_values = iris[numerical_columns].mean()
median_values = iris[numerical_columns].median()
mode_values = iris[numerical_columns].mode().iloc[0]
stddev_values = iris[numerical_columns].std()

print("\nMean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nMode Values:")
print(mode_values)
print("\nStandard Deviation Values:")
print(stddev_values)
```
